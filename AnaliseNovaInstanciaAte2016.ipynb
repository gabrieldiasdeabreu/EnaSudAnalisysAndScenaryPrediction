```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
baseDados = np.loadtxt('Instancia/EnaSudNovoAte2016.csv',delimiter=',')
meses = ('janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho', 'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro')
# colunasBase = (list(x) for x in baseDados)
# baseDados = {meses[x]:colunasBase[x] for x in range(13)}
baseDados = pd.DataFrame(baseDados,columns=meses)
baseDados
```

```python
baseDados.T.plot(title="Todos os anos EnaSud", legend=False,figsize=[20,10])
```

```python
baseDados.plot(kind='box',title="Todos os anos EnaSud", legend=False, figsize=[20,10])
```

# separacao do treino Validacao e Teste


```python
tamTreino = int(np.ceil(len(baseDados)*0.80))
tamValidacao = int(len(baseDados)*0.10)
tamTeste = int((len(baseDados)*0.10))
print('tamanho do conjunto:')
print('Treino:', tamTreino)
print('Validacao:', tamValidacao)
print('Teste:', tamTeste)
print('Total repartido:', tamTeste+tamTreino+tamValidacao, ' de ', len(baseDados))
print('split', tamValidacao/(tamTreino+tamValidacao))
treino = baseDados[:tamTreino]
validacao = baseDados[tamTreino:tamTreino+tamValidacao]
teste = baseDados[tamTreino+tamValidacao:]
treino.boxplot(figsize=[20,10])
```
